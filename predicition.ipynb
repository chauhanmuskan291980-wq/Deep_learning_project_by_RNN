```python
from tensorflow.keras.models import load_model

model = load_model('/content/drive/MyDrive/simple_rnn-imdb.h5')
```

```python
model.summary()
```

```python
word_index = imdb.get_word_index()
reverse_word__index = {value:key for key , value in word_index.items()}
```

```python
model.get_weights()
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

max_len = 500

def preprocess_text(text):
    words = text.lower().split()
    encoded_review = [word_index.get(word, 2) + 3 for word in words]
    padded_review = pad_sequences(
        [encoded_review],
        maxlen=max_len
    )
    return padded_review
```

```python
def predict_sentences(reviews):
  preprocesed_text=preprocess_text(reviews)
  prediction = model.predict(preprocesed_text)
  sentiment = 'Positive' if prediction[0][0] > 0.5 else 'Negative'

  return sentiment , prediction[0][0]
```

```python
example_review = 'This movie was fantastic! The acting was great and the plot was thrilling'

sentiment , score = predict_sentences(example_review)
print(sentiment,score)
```
